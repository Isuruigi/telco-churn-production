```python
# üìä TELCO CHURN - COMPREHENSIVE DATA EXPLORATION
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
import warnings
warnings.filterwarnings('ignore')

print("üîç TELCO CUSTOMER CHURN - DATA EXPLORATION")
print("=" * 60)

# ==========================================
# SECTION 1: DATA LOADING & BASIC INFO
# ==========================================
print("\nüì• SECTION 1: Loading and Basic Information")
print("-" * 50)

try:
    df = pd.read_csv('../data/WA_Fn-UseC_-Telco-Customer-Churn.csv')
    print("‚úÖ Dataset loaded successfully!")
    
    print(f"\nüìä Dataset Overview:")
    print(f"   ‚Ä¢ Shape: {df.shape}")
    print(f"   ‚Ä¢ Total Customers: {len(df):,}")
    print(f"   ‚Ä¢ Features: {len(df.columns)}")
    print(f"   ‚Ä¢ Memory Usage: {df.memory_usage(deep=True).sum() / 1024**2:.2f} MB")
    
    print(f"\nüìã Column Names:")
    print(df.columns.tolist())
    
except FileNotFoundError:
    print("‚ùå Dataset not found!")
    print("üì• Please download from Kaggle and save as: data/WA_Fn-UseC_-Telco-Customer-Churn.csv")

if 'df' not in locals():
    raise Exception("Cannot proceed without dataset")

# ==========================================
# SECTION 2: DATA STRUCTURE ANALYSIS
# ==========================================
print(f"\nüîç SECTION 2: Data Structure Analysis")
print("-" * 45)

# Data types and missing values
print("üìã Data Types and Missing Values:")
info_summary = pd.DataFrame({
    'Column': df.columns,
    'Data_Type': df.dtypes,
    'Non_Null_Count': df.count(),
    'Null_Count': df.isnull().sum(),
    'Null_Percentage': (df.isnull().sum() / len(df)) * 100,
    'Unique_Values': [df[col].nunique() for col in df.columns],
    'Sample_Values': [str(df[col].unique()[:3])[1:-1] for col in df.columns]
})
display(info_summary)

# Separate numerical and categorical columns
numerical_cols = df.select_dtypes(include=[np.number]).columns.tolist()
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()

print(f"\nüìä Column Categories:")
print(f"   ‚Ä¢ Numerical columns ({len(numerical_cols)}): {numerical_cols}")
print(f"   ‚Ä¢ Categorical columns ({len(categorical_cols)}): {categorical_cols}")

# ==========================================
# SECTION 3: TARGET VARIABLE ANALYSIS
# ==========================================
print(f"\nüéØ SECTION 3: Target Variable Analysis")
print("-" * 45)

churn_distribution = df['Churn'].value_counts()
churn_percentage = df['Churn'].value_counts(normalize=True) * 100

print("üìà Churn Distribution:")
for status, count in churn_distribution.items():
    percentage = churn_percentage[status]
    print(f"   ‚Ä¢ {status}: {count:,} customers ({percentage:.1f}%)")

# ==========================================
# SECTION 4: COMPREHENSIVE VISUALIZATIONS
# ==========================================
print(f"\nüìä SECTION 4: Creating Comprehensive Visualizations")
print("-" * 55)

# Set up the plotting style
plt.style.use('default')
sns.set_palette("Set2")

# Figure 1: Basic Distribution Analysis
fig1, axes = plt.subplots(2, 3, figsize=(20, 12))
fig1.suptitle('Telco Churn Analysis - Basic Distributions', fontsize=16, fontweight='bold')

# Plot 1: Churn Distribution
churn_distribution.plot(kind='bar', ax=axes[0,0], color=['skyblue', 'salmon'])
axes[0,0].set_title('Customer Churn Distribution', fontweight='bold')
axes[0,0].set_xlabel('Churn Status')
axes[0,0].tick_params(axis='x', rotation=0)
for i, v in enumerate(churn_distribution.values):
    axes[0,0].text(i, v + 50, str(v), ha='center', va='bottom', fontweight='bold')

# Plot 2: Gender Distribution
df['gender'].value_counts().plot(kind='pie', ax=axes[0,1], autopct='%1.1f%%', colors=['lightblue', 'lightpink'])
axes[0,1].set_title('Gender Distribution', fontweight='bold')
axes[0,1].set_ylabel('')

# Plot 3: Senior Citizen Distribution
df['SeniorCitizen'].value_counts().plot(kind='bar', ax=axes[0,2], color=['lightgreen', 'orange'])
axes[0,2].set_title('Senior Citizen Distribution', fontweight='bold')
axes[0,2].set_xlabel('Senior Citizen (0=No, 1=Yes)')
axes[0,2].tick_params(axis='x', rotation=0)

# Plot 4: Tenure Distribution
df['tenure'].hist(bins=30, ax=axes[1,0], color='purple', alpha=0.7, edgecolor='black')
axes[1,0].set_title('Customer Tenure Distribution', fontweight='bold')
axes[1,0].set_xlabel('Tenure (months)')
axes[1,0].set_ylabel('Number of Customers')

# Plot 5: Monthly Charges Distribution
df['MonthlyCharges'].hist(bins=30, ax=axes[1,1], color='orange', alpha=0.7, edgecolor='black')
axes[1,1].set_title('Monthly Charges Distribution', fontweight='bold')
axes[1,1].set_xlabel('Monthly Charges ($)')
axes[1,1].set_ylabel('Number of Customers')

# Plot 6: Total Charges Distribution (handle non-numeric values)
df_clean = df.copy()
df_clean['TotalCharges'] = pd.to_numeric(df_clean['TotalCharges'], errors='coerce')
df_clean['TotalCharges'].dropna().hist(bins=30, ax=axes[1,2], color='green', alpha=0.7, edgecolor='black')
axes[1,2].set_title('Total Charges Distribution', fontweight='bold')
axes[1,2].set_xlabel('Total Charges ($)')
axes[1,2].set_ylabel('Number of Customers')

plt.tight_layout()
plt.show()

# Figure 2: Service and Contract Analysis
fig2, axes = plt.subplots(2, 3, figsize=(20, 12))
fig2.suptitle('Service Usage and Contract Analysis', fontsize=16, fontweight='bold')

# Contract types
df['Contract'].value_counts().plot(kind='bar', ax=axes[0,0], color='lightcoral')
axes[0,0].set_title('Contract Type Distribution', fontweight='bold')
axes[0,0].tick_params(axis='x', rotation=45)

# Payment Methods
df['PaymentMethod'].value_counts().plot(kind='bar', ax=axes[0,1], color='gold')
axes[0,1].set_title('Payment Method Distribution', fontweight='bold')
axes[0,1].tick_params(axis='x', rotation=45)

# Internet Service
df['InternetService'].value_counts().plot(kind='pie', ax=axes[0,2], autopct='%1.1f%%')
axes[0,2].set_title('Internet Service Distribution', fontweight='bold')
axes[0,2].set_ylabel('')

# Phone Service
df['PhoneService'].value_counts().plot(kind='bar', ax=axes[1,0], color='lightblue')
axes[1,0].set_title('Phone Service Distribution', fontweight='bold')
axes[1,0].tick_params(axis='x', rotation=0)

# Multiple Lines
df['MultipleLines'].value_counts().plot(kind='bar', ax=axes[1,1], color='lightgreen')
axes[1,1].set_title('Multiple Lines Distribution', fontweight='bold')
axes[1,1].tick_params(axis='x', rotation=45)

# Paperless Billing
df['PaperlessBilling'].value_counts().plot(kind='pie', ax=axes[1,2], autopct='%1.1f%%', colors=['wheat', 'lightcyan'])
axes[1,2].set_title('Paperless Billing Distribution', fontweight='bold')
axes[1,2].set_ylabel('')

plt.tight_layout()
plt.show()

# ==========================================
# SECTION 5: CHURN ANALYSIS BY FEATURES
# ==========================================
print(f"\nüìà SECTION 5: Churn Analysis by Key Features")
print("-" * 50)

# Create churn analysis visualizations
fig3, axes = plt.subplots(2, 3, figsize=(20, 12))
fig3.suptitle('Churn Analysis by Key Features', fontsize=16, fontweight='bold')

# Churn by Contract
churn_contract = pd.crosstab(df['Contract'], df['Churn'], normalize='index') * 100
churn_contract.plot(kind='bar', ax=axes[0,0], color=['lightblue', 'salmon'])
axes[0,0].set_title('Churn Rate by Contract Type', fontweight='bold')
axes[0,0].set_ylabel('Percentage (%)')
axes[0,0].legend(['No Churn', 'Churn'])
axes[0,0].tick_params(axis='x', rotation=45)

# Churn by Payment Method
churn_payment = pd.crosstab(df['PaymentMethod'], df['Churn'], normalize='index') * 100
churn_payment.plot(kind='bar', ax=axes[0,1], color=['lightblue', 'salmon'])
axes[0,1].set_title('Churn Rate by Payment Method', fontweight='bold')
axes[0,1].set_ylabel('Percentage (%)')
axes[0,1].legend(['No Churn', 'Churn'])
axes[0,1].tick_params(axis='x', rotation=45)

# Churn by Internet Service
churn_internet = pd.crosstab(df['InternetService'], df['Churn'], normalize='index') * 100
churn_internet.plot(kind='bar', ax=axes[0,2], color=['lightblue', 'salmon'])
axes[0,2].set_title('Churn Rate by Internet Service', fontweight='bold')
axes[0,2].set_ylabel('Percentage (%)')
axes[0,2].legend(['No Churn', 'Churn'])
axes[0,2].tick_params(axis='x', rotation=45)

# Tenure vs Churn
df.boxplot(column='tenure', by='Churn', ax=axes[1,0])
axes[1,0].set_title('Tenure Distribution by Churn Status', fontweight='bold')
axes[1,0].set_xlabel('Churn Status')
axes[1,0].set_ylabel('Tenure (months)')

# Monthly Charges vs Churn
df.boxplot(column='MonthlyCharges', by='Churn', ax=axes[1,1])
axes[1,1].set_title('Monthly Charges by Churn Status', fontweight='bold')
axes[1,1].set_xlabel('Churn Status')
axes[1,1].set_ylabel('Monthly Charges ($)')

# Correlation heatmap for numerical features
numerical_df = df.select_dtypes(include=[np.number])
correlation_matrix = numerical_df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, ax=axes[1,2])
axes[1,2].set_title('Correlation Matrix - Numerical Features', fontweight='bold')

plt.tight_layout()
plt.show()

# ==========================================
# SECTION 6: DATA QUALITY ASSESSMENT
# ==========================================
print(f"\nüîç SECTION 6: Data Quality Assessment")
print("-" * 45)

# Check for data quality issues
print("üîé Data Quality Issues:")

# Check for missing values
missing_values = df.isnull().sum().sum()
print(f"   ‚Ä¢ Missing Values: {missing_values}")

# Check for duplicate rows
duplicates = df.duplicated().sum()
print(f"   ‚Ä¢ Duplicate Rows: {duplicates}")

# Check for inconsistent data types (TotalCharges should be numeric)
print(f"   ‚Ä¢ TotalCharges Data Type Issues:")
total_charges_issues = df['TotalCharges'].apply(lambda x: not str(x).replace('.', '').isdigit() if x != ' ' else True).sum()
print(f"     - Non-numeric TotalCharges: {total_charges_issues}")

# Check unique values for binary columns
print(f"\nüìã Binary Column Unique Values:")
binary_expected = ['Yes', 'No']
binary_cols = ['Partner', 'Dependents', 'PhoneService', 'PaperlessBilling', 'Churn']
for col in binary_cols:
    unique_vals = df[col].unique().tolist()
    is_binary = set(unique_vals).issubset(set(binary_expected))
    print(f"   ‚Ä¢ {col}: {unique_vals} ({'‚úÖ Binary' if is_binary else '‚ùå Not Binary'})")

# ==========================================
# SECTION 7: KEY INSIGHTS & RECOMMENDATIONS
# ==========================================
print(f"\nüí° SECTION 7: Key Insights & Recommendations")
print("-" * 55)

# Calculate key metrics
churn_rate = (df['Churn'] == 'Yes').mean() * 100
high_value_customers = df[df['MonthlyCharges'] > df['MonthlyCharges'].quantile(0.8)]
high_value_churn_rate = (high_value_customers['Churn'] == 'Yes').mean() * 100

print("üîç Key Business Insights:")
print(f"   ‚Ä¢ Overall churn rate: {churn_rate:.1f}%")
print(f"   ‚Ä¢ High-value customer churn rate: {high_value_churn_rate:.1f}%")
print(f"   ‚Ä¢ Average tenure: {df['tenure'].mean():.1f} months")
print(f"   ‚Ä¢ Average monthly charges: ${df['MonthlyCharges'].mean():.2f}")

# Contract analysis
contract_churn = df.groupby('Contract')['Churn'].apply(lambda x: (x == 'Yes').mean() * 100)
print(f"\nüìã Churn Rate by Contract Type:")
for contract, rate in contract_churn.items():
    print(f"   ‚Ä¢ {contract}: {rate:.1f}%")

print(f"\nüéØ Recommendations for ML Pipeline:")
print("   1. Handle 'TotalCharges' data type conversion")
print("   2. Encode categorical variables appropriately") 
print("   3. Consider feature engineering for tenure groups")
print("   4. Address class imbalance if needed")
print("   5. Focus on contract type as key predictor")
print("   6. Consider customer lifetime value in model")

print(f"\n‚úÖ Data exploration completed successfully!")
print(f"üöÄ Ready to proceed with Part 1: Scikit-learn Pipeline Development")

# Save cleaned dataset for pipeline development
print(f"\nüíæ Preparing cleaned dataset for pipeline...")
df_for_pipeline = df.copy()
df_for_pipeline['TotalCharges'] = pd.to_numeric(df_for_pipeline['TotalCharges'], errors='coerce')
df_for_pipeline['TotalCharges'].fillna(df_for_pipeline['TotalCharges'].median(), inplace=True)
df_for_pipeline.to_csv('../data/processed/telco_churn_cleaned.csv', index=False)
print("‚úÖ Cleaned dataset saved as: ../data/processed/telco_churn_cleaned.csv")
```
